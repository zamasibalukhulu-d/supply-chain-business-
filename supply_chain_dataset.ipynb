```python
# Looking at the revenue per product to see what it contains the most revenue
revbyproduct = data.groupby('Product type')['Revenue generated'].sum()
print(revbyproduct)
```

```python
quantity_sold = data.groupby(['Order quantities', 'Number of products sold'])(['Product type'].unique()
print(quantity_sold)
```

```python
trans_cost = data.groupby('Transportation modes')['Costs'].sum()
print(trans_cost)
```

```python
ship = data.groupby('Shipping carriers')['Shipping costs'].sum()
print(ship)
```

```python
ship_2 = data.groupby('Shipping carriers')['Product type'].value_counts()
print(ship_2)
```

```python
ship_2_unstacked = ship_2.unstack()
ship_2_unstacked.plot(kind='bar', figsize=(10, 6))
plt.title('Product Types Shipped by Carrier')
plt.xlabel('Shipping Carrier')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.legend(title='Product Type')
plt.tight_layout()
plt.show()
```

```python
colors = ['lightcoral', 'lightskyblue', 'lightgreen']
plt.pie(revbyproduct, colors=colors, labels=revbyproduct.index, autopct='%1.1f%%', startangle=90)
plt.title('Revenue Made By The Products')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
```

```python
plt.scatter(data['Manufacturing costs'], data['Price'])
plt.title('Manufacturing cost Vs Price')
plt.xlabel('Manufacturing Cost')
plt.ylabel('Price sold by')
# set the range
plt.xlim(0, 100)  # 100 to 300
plt.ylim(46, 100)  #100 to 1,000
```
